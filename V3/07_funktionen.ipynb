```python
'''    
# Aufgabe 1

Erstelle eine Funktion func1, der eine Zahl k übergeben wird und 
die das Quadrat dieser Zahl zurückgibt.

>>> func1(4)
16

>>> func1(5)
25
'''
def func1(k):
    return k * k
```
