### Matplotlib

Matplotlib ist eine Python-Bibliothek mit dem wir Plots zur Visualisierung von Daten erzeugen können.

Installation mit: ```pip install matplotlib```

Meistens nutzen wir Funktionen aus dem pyplot submodule, das üblicherweise mit dem alias plt importiert wird.

````python
import matplotlib.pyplot as plt
````

#### Punkte plotten

Mit der Funktion *scatter()* können wir Punkte plotten.
Die x-Koordinaten und y-Koordinaten werden als Listen eingegeben. Die Parameter für Dicke oder Farbe können entweder für alle Punkte
oder mit einer Liste für jeden Punkt spezifiziert werden.    

````python
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
plt.scatter(X, Y, color='red', s=30, marker='o')        # s = dicke
plt.show()
````

Als Farben können wir die üblichen englischen Namen für Farben verwenden. <br>
Als Marker können wir verwenden, z.B:

    'o'  für dicke Punkte
    '.'  für dünne Punkte
    'x'  für eine x-Markierung

#### Größe des Plots

Die Größe des Plots wird über die Parameter in *figsize* bestimmt: Breite und Höhe in Inch (1 Inch = 2.54 cm). Das ist die Größe, wenn wir das Bild mit *savefig()* speichern. Die angezeigte Größe hängt u.a. von der Bildschirmauflösung ab. Meist passen wir die Bildgröße durch Ausprobieren für eine konkreten Fall an. 

````python
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
C = ['red','red','blue','blue']

plt.figure(figsize=(4,3))                           # 4 inch breit, 3 inch hoch 
plt.scatter(X, Y, color=C, s=30, marker='o')        # s = dicke

plt.show()
````

Die Markierungen der Punkte und die Farben können wir ändern

#### Linien zwischen Punkten

Die Funktion *plot* zeichnet Linien zwischen den Punkten.   

````python
X = [1, 8, 7, 3]
Y = [3, 10, 2, 9]

plt.figure(figsize=(4,3))      
plt.plot(X, Y)
plt.show()
````

Der Parameter *linestyle* bestimmt die Art der Linie, z.B:

    '-'          solid line style (default)
    '--'         dashed line style
    '-.'         dash-dot line style
    ':'          dotted line style

````python
X = [1, 8, 7, 3]
Y = [3, 10, 2, 9]

plt.figure(figsize=(4,3))      
plt.plot(X, Y, color='red', linestyle='--', marker="o")
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3*x**2 - x + 3

def f1(x):
    return 3*x**2 - 6*x - 1

X = np.linspace(-2, 4, 400)
Y = f(X)
Y1 = f1(X)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(X, Y, label='x**3 - 3*x**2 - x + 3', color="blue")
plt.plot(X, Y1, label='3*x**2 - 6*x - 1', color="red")

# Beschriftung
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title("Polynom mit Ableitung")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
````

Wir können die Grenzen für x- und y-Achse auch selbst vorgeben. Hier sind zusätzlich die Nullstellen der Ableitung (und damit die Extremstellen) des Polynoms mit einer schwachen Linie markiert.

````python
import numpy as np
import matplotlib.pyplot as plt
# Plot
plt.figure(figsize=(10, 6))
plt.plot(X, Y, label='x**3 - 3*x**2 - x + 3', color="blue")
plt.plot(X, Y1, label='Ableitung', color="red")
          
plt.ylim(-5, 4)     
plt.xlim(-2, 4)  
 
# Beschriftung

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.axvline(-0.155, color='green', linewidth=0.8, linestyle=':')
plt.axvline(2.155, color='green', linewidth=0.8, linestyle=':')

plt.title("Polynom mit Ableitung")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt

# Liste der Punkte
points = [(-4,1), (-3,3), (-5,2), (-2,2), (-6,-2), (-4,-2), (1,1), (1,-1), (7,5)]

# X- und Y-Koordinaten extrahieren
x_coords = [x for x, y in points]
y_coords = [y for x, y in points]

# Figur und Achsenobjekt
fig, ax = plt.subplots()

# Punkte zeichnen
ax.scatter(x_coords, y_coords, color='blue')

# Punkte beschriften
for i, (x, y) in enumerate(points):
    ax.text(x + 0.1, y + 0.1, f'P{i}', fontsize=9)

# Gitter einblenden (dünn)
ax.grid(True, linewidth=0.5, color='gray', linestyle='--')

# Achsen durch Ursprung (als dicker Pfeil dargestellt)
ax.axhline(0, color='black', linewidth=1.5)
ax.axvline(0, color='black', linewidth=1.5)

# Rahmen entfernen
for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

# Ticks an den Achsen positionieren
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.tick_params(axis='x', direction='inout', length=5)
ax.tick_params(axis='y', direction='inout', length=5)

# Ticks direkt an den Achsen positionieren
ax.xaxis.set_label_coords(1.02, 0.5)
ax.yaxis.set_label_coords(0.5, 1.02)

# Achsen gleich skalieren
ax.set_aspect('equal')

# Achsen beschriften
ax.set_xlabel("x", labelpad=-20)
ax.set_ylabel("y", labelpad=-25)
ax.set_title("Punkte im Koordinatensystem")

# Anzeigen
plt.show()

````

````python
import matplotlib.pyplot as plt

# Punkte
points = [(-4,1), (-3,3), (-5,2), (-2,2), (-6,-2), (-4,-2), (1,1), (1,-1), (7,5)]
x_coords = [x for x, y in points]
y_coords = [y for x, y in points]

# Plotbereich erstellen
fig, ax = plt.subplots()
ax.set_aspect('equal')

# Punkte zeichnen
ax.scatter(x_coords, y_coords, color='blue')

# Punkte beschriften
for i, (x, y) in enumerate(points):
    ax.text(x + 0.2, y + 0.2, f'P{i}', fontsize=9)

# Grenzen setzen mit etwas Puffer
xmin, xmax = min(x_coords)-1, max(x_coords)+1
ymin, ymax = min(y_coords)-1, max(y_coords)+1
ax.set_xlim(xmin, xmax)
ax.set_ylim(ymin, ymax)

# Achsenpfeile zeichnen
arrowprops = dict(arrowstyle='->', linewidth=1.8, color='black')
ax.annotate('', xy=(xmax, 0), xytext=(xmin, 0), arrowprops=arrowprops)  # x-Achse
ax.annotate('', xy=(0, ymax), xytext=(0, ymin), arrowprops=arrowprops)  # y-Achse

# Ticks setzen
xticks = range(xmin+1, xmax)
yticks = range(ymin+1, ymax)
ax.set_xticks(xticks)
ax.set_yticks(yticks)

# Achsenticks anpassen: direkt an der Achse positionieren
ax.tick_params(axis='both', which='both', length=0)  # keine Tick-Striche

# Tick-Labels direkt an der Achse platzieren
for x in xticks:
    if x != 0:
        ax.text(x, -0.4, str(x), ha='center', va='top', fontsize=9)
for y in yticks:
    if y != 0:
        ax.text(-0.4, y, str(y), ha='right', va='center', fontsize=9)

# Ursprung beschriften
ax.text(0.3, 0.3, "0", fontsize=9)

# Rahmen entfernen und Gitter anzeigen
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(True, linewidth=0.5, linestyle='--', color='gray')

# Achsenbeschriftungen
ax.set_title("Punkte im Koordinatensystem")

# Achsenbeschriftungspfeile ("x" und "y")
ax.text(xmax, -0.3, "x", fontsize=12, ha='left')
ax.text(0.3, ymax, "y", fontsize=12, va='bottom')

# Anzeige
plt.show()
````
