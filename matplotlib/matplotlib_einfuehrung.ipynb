In diesem Notebook lernst du, wie man mit `matplotlib.pyplot` Daten in Python grafisch darstellen kann.


```python
import matplotlib.pyplot as plt
```

## üìà 1. Einfache Liniengrafik

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title("Quadratzahlen")
plt.xlabel("x-Werte")
plt.ylabel("y-Werte")
plt.grid(True)
plt.show()
```

## üé® 2. Farben, Linienarten und Marker

```python
plt.plot(x, y, color="red", linestyle="--", marker="o")
plt.title("Gestylte Linie")
plt.show()
```

| Parameter       | Beispiel         | Bedeutung               |
|----------------|------------------|-------------------------|
| `color`        | `"blue"`         | Farbe                   |
| `linestyle`    | `"-"`, `"--"`    | Linie durchgezogen/gestrichelt |
| `marker`       | `"o"`, `"x"`     | Punktmarkierungen       |


## üìä 3. Balkendiagramm (Bar Chart)

```python
namen = ["Anna", "Ben", "Clara", "David"]
punkte = [15, 18, 12, 20]

plt.bar(namen, punkte)
plt.title("Punktestand")
plt.ylabel("Punkte")
plt.show()
```

## üßÅ 4. Kreisdiagramm (Pie Chart)

```python
labels = ["Python", "Java", "C++", "JavaScript"]
sizes = [40, 25, 20, 15]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Programmiersprachen-Beliebtheit")
plt.show()
```

## üìö 5. Mehrere Linien in einem Diagramm

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, label="Linear")
plt.plot(x, y2, label="Quadratisch")
plt.legend()
plt.title("Zwei Funktionen")
plt.show()
```

## üîÑ 6. Diagramm speichern

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.savefig("mein_diagramm.png")
```

## ‚ÑπÔ∏è Zusammenfassung ‚Äì Wichtige Befehle

| Funktion             | Beschreibung                        |
|----------------------|-------------------------------------|
| `plt.plot()`         | Liniengrafik                        |
| `plt.bar()`          | Balkendiagramm                      |
| `plt.pie()`          | Kreisdiagramm                       |
| `plt.title()`        | Titel setzen                        |
| `plt.xlabel()`, `plt.ylabel()` | Achsenbeschriftung         |
| `plt.legend()`       | Legende anzeigen                    |
| `plt.grid()`         | Gitternetz ein-/ausschalten         |
| `plt.savefig()`      | Grafik speichern                    |
| `plt.show()`         | Grafik anzeigen                     |

```python
import numpy as np
import matplotlib.pyplot as plt

# Definition der Funktion f(x)
def f(x):
    return x**3 - 3*x**2 - x + 3

# Erste und zweite Ableitung
def f_prime(x):
    return 3*x**2 - 6*x - 1

def f_double_prime(x):
    return 6*x - 6

# x-Werte
x = np.linspace(-2, 4, 400)
y = f(x)

# Nullstellen (ungef√§hr, hier zur Vereinfachung vorgegeben)
nullstellen_x = [-1, 1, 3]
nullstellen_y = [f(xi) for xi in nullstellen_x]

# Extrempunkte (ungef√§hr per Rechnung oder CAS gefunden)
extrema_x = [1 - np.sqrt(10)/3, 1 + np.sqrt(10)/3]
extrema_y = [f(xi) for xi in extrema_x]

# Wendepunkt (f''(x) = 0 ‚Üí x = 1)
wendepunkt_x = [1]
wendepunkt_y = [f(1)]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = x¬≥ - 3x¬≤ - x + 3", color="blue")

# Nullstellen
plt.scatter(nullstellen_x, nullstellen_y, color="red", label="Nullstellen")

# Extrempunkte
plt.scatter(extrema_x, extrema_y, color="green", label="Extrempunkte")

# Wendepunkt
plt.scatter(wendepunkt_x, wendepunkt_y, color="orange", label="Wendepunkt")

# Beschriftung
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title("Polynom mit Nullstellen, Extrempunkten und Wendepunkt")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
