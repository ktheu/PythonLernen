### Dictionaries

```python
m = {}     # ein leeres dictionary
m = {'e' : 10, 'n' : 20, 'i' : 30}    # ein dict mit 3 key-value paaren
```

```python
# durch ein dict laufen und die key-value Paare ausgeben
for k in m:
    print(k, m[k])
```

##### Beispiel 

```python
# Für die Buchstaben e, n und i erhalten wir Punkte (wie im dict gespeichert)
# Wir wollen die Summe aller Punkte in einem Text ermitteln
m = {'e' : 1, 'n' : 2, 'i' : 3}  
s = 'In der Innenwelt des Computers werden Nullen und Einsen verschoben'
s = s.lower()                     # alles zu Kleinbuchstaben
punkte = 0
for c in s:                       # für jedes Zeichen c in s
    if c in m:                    # falls c als key in m vorkommt
        punkte = punkte + m[c]    # erhöhe die Anzahl der Punkte 
print(punkte)
    
```

### Aufgaben




```python
'''
Aufgabe 1:
Gegeben ist ein text und ein dictionary m, das einigen Zeichen Zahlen zuordnet.
Schreibe ein Programm das zählt, wieviele Zeichen von s im dictionary m als key vorkommen.

text = 'In der Innenwelt des Computers werden Nullen und Einsen verschoben, verrechnet und gespeichert'
m = {'l': 1, 'e': 3, 'd': 2, 'p': 2, 'c': 1, 'h': 3, 'b': 3, 'j': 2, 'u': 1}

Erwartete Ausgabe
38
'''
text = 'In der Innenwelt des Computers werden Nullen und Einsen verschoben, verrechnet und gespeichert'
m = {'l': 1, 'e': 3, 'd': 2, 'p': 2, 'c': 1, 'h': 3, 'b': 3, 'j': 2, 'u': 1}

zaehl = 0
for c in text:
    if c in m:
        zaehl = zaehl + 1
print(zaehl)
```

```python
'''
Aufgabe 2
Gegeben ist eine Liste von Zahlen. Wenn eine Zahl durch 2 teilbar ist, erhalten wir 1 Punkt, ist sie
durch 3 oder 5 teilbar, erhalten wir 2 Punkte, ist sie durch 7 teilbar, erhalten wir 5 Punkte, 
ist sie durch 11 teilbar, erhalten wir 7 Punkte. 
Ermittle die Punkte für die Liste a = [2,3,5,1,11]

Pseudocode:

Erstelle ein dictionary m, das der Teilungszahl seine Punktezahl zuordnet: 2->1, 3->2, 5->2, 7->5, 11->7

setze punkte auf 0
für jedes x in a:
     wenn x key in m:
         erhöhe punkte um den Wert, der x in m zugeordnet ist
gib punkte aus

Erwartete Ausgabe:
12
'''
a = [2,3,5,1,11]
m = {2: 1, 3: 2, 5: 2, 7: 5, 11: 7}

punkte = 0
for x in a:
    for k in m:
        if x % k == 0:
            punkte = punkte + m[k]
print(punkte)
```

```python
'''
Aufgabe 3
s = 'In der Innenwelt des Computers werden Nullen und Einsen verschoben'

Wir beginnen beim ersten Zeichen des Strings s. Wir ermitteln den ASCII-Wert modulo 7 des Zeichens und erhöhen
diesen Wert um 1. Soviele Schritte rücken wir im String vor. Wir zählen, wie häufig wir das machen müssen, bis wir
außerhalb des Strings landen.

Pseudocode: 

Setze pos auf 0
Setze zaehl auf 0
Solange pos kleiner als die Länge von s:
    Erhöhe zaehl um 1
    Setze c auf das Zeichen von s an der Stelle pos.
    Setze w auf den ASCII-Wert von c modulo 7
    Erhöhe pos um w + 1
Gib zaehl aus


Erwartete Ausgabe: 15
'''

s = 'In der Innenwelt des Computers werden Nullen und Einsen verschoben'
pos = 0
zaehl = 0
while pos < len(s):
    zaehl += 1
    c = s[pos]
    weiter = ord(c) % 7
    pos = pos + 1 + weiter
print(zaehl) 
    
    
```
