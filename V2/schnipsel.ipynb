```python
s = 'Habicht'
for i in range(len(s)):             # durch die Zeichen eines Strings mit dem Index laufen.
    print(s[i])

s = 'Habicht'
for c in s:                         # durch die Zeichen eines Strings laufen
    print(c)
```

##### 05

```python
# Slicing (Teilstrings bilden)
s = 'Habicht'
s[:3]        # 'Hab'
s[3:]        # 'icht'
s[2:5]       # 'bic'
s[::2]       # jedes zweite Zeichen: 'Hbct'
s[::-1]      # den gesamten String rückwärts:  'thcibaH'
```

##### 06

```python
# Listen
a = []        # leere Liste
a = [1,2,3]   # eine Liste mit 3 Elementen
len(a)        # Länge der Liste
a[0]          # 1    Indexing wie bei Strings
a[len(a)-1]   # 3
a[-1]         # 3         
a.append(4)   # [1,2,3,4]   ein Element hinten an die Liste einfügen.
```

```python
# Teillisten ähnlich wie Teilstrings 
a = [6,3,8,2,1,4]
a[:2]    # [6,3]          
a[2:]    # [8,2,1,4]      
a[2:5]   # [8,2,1]         
a[::-1]  # [4,1,2,8,3,6]
```

```python
# input mit vielen Zahlen
s = input('Bitte eine Folge von Zahlen eingeben:')
a = []
for x in s.split():          # Diese Art der for-Schleife betrachten wir später genauer
    a.append(int(x))
print(a)
```
